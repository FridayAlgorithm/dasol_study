## 3085 사탕게임

**문제로 이동** : [Baekjoon 사탕게임](https://www.acmicpc.net/problem/3085)

- 처음 N\*N 크기에 사탕을 채워 넣는다
- 사탕이 색이 다른 인접한 두 칸의 사탕을 교환한다
- 같은 색으로 이루어져있는 연속 부분을 모두 먹는다
- 사탕이 채워진 상태가 주어질 때 먹을 수 있는 사탕의 최대 개수를 구하라

```python
N=int(input())
candyColor=[list(input()) for _ in range(N)]

def getCandyNum(x, y):
    row, col, maxCount=0,0,0
    rowColor=candyColor[0][y]
    colColor=candyColor[x][0]
    for i in range(N):
        if candyColor[i][y]==rowColor:
            row+=1
            if maxCount<row:
                maxCount=row
        else:
            rowColor=candyColor[i][y]
            row=1
        
        if candyColor[x][i]==colColor:
            col+=1
            if maxCount<col:
                maxCount=col
        else:
            colColor=candyColor[x][i]
            col=1
    return maxCount

    
maxCount=[]
for i in range(N):
    for j in range(N):
        if j<N-1:
            candyColor[i][j],candyColor[i][j+1]=candyColor[i][j+1],candyColor[i][j]
            maxCount.append(getCandyNum(i,j))
            maxCount.append(getCandyNum(i,j+1))
            candyColor[i][j+1],candyColor[i][j]=candyColor[i][j],candyColor[i][j+1]
        if i<N-1:
            candyColor[i][j],candyColor[i+1][j]=candyColor[i+1][j],candyColor[i][j]
            maxCount.append(getCandyNum(i,j))
            maxCount.append(getCandyNum(i+1,j))
            candyColor[i+1][j],candyColor[i][j]=candyColor[i][j],candyColor[i+1][j]

print(max(maxCount))
```

## 3135 라디오

**문제로 이동** : [Baekjoon 라디오](https://www.acmicpc.net/problem/3135)

- 라디오는 다음과 같은 버튼이 있다
  * 첫번째 버튼은 주파수를 1MHz 증가시킨다
  * 두번째 버튼은 주파수를 1MHz 감소시킨다
  * 나머지 N개의 버튼은 즐겨찾기 기능으로 미리 지정된 주파수로 이동한다
- 현재 주파수 A와 듣고싶은 주파수 B가 주어질 때 A에서 B로 이동하는 가장 적은 횟수의 버튼수를 구하자

```python
A, B=map(int,input().split())
N=int(input())
frequency=[int(input()) for _ in range(N)]

minCount=abs(B-A)
result=0
for f in frequency:
    if (abs(B-f)<minCount):
        result=1
        minCount=abs(B-f)
result+=minCount
print(result)
```

## 4307 개미

**문제로 이동** : [Baekjoon 개미](https://www.acmicpc.net/problem/4307)

- 길이가 lcm인 막대 위에 개미 여러마리가 있다. 
- 개미의 속도는 1cm/s로 일정하다
- 개미는 마지막에 도달하면 떨어지고, 다른 개미와 만나면 서로 방향을 바꾼다
- 개미가 땅으로 떨어질 떄까지의 가장 빠른 시간과 느린 시간을 구하라

```python
#시간초과 -> pypy
testNum=int(input())
for _ in range(testNum):
    bar, ant=map(int,input().split())
    locate=[int(input()) for _ in range(ant)]
    minTime, maxTime=0, 0
    for i in locate:
        temp=min(i, bar-i)
        minTime=max(minTime, temp)
        maxTime=max(maxTime, i, bar-i)
    print(minTime, maxTime)
```

## 17135 캐슬 디펜스

**문제로 이동** : [Baekjoon 캐슬 디펜스](https://www.acmicpc.net/problem/17135)

- 크기가 N\*M인 격자판에서 적을 잡는 턴 방식의 게임을 진행한다
- 각 칸의 최대 적의 수는 하나이다
- N번행 바로 아래의 모든 칸에는 성이 있다
- 3명의 궁수를 배치하는데, 궁수는 거리가 D 이하인 적 중 가장 가까운 적을 공격한다
- 궁수의 공격이 끝나면 적이 아래로 한 칸 이동하고, 성의 칸의 도달하면 게임에서 제외된다
- 모든 적이 격자판에서 제외되면 게임이 끝난다
- 격자판의 상태가 주어질 때 궁수가 제거할 수 있는 적의 최대 수를 구하라

```python
from collections import deque

n, m, d=map(int,input().split())
board=[list(map(int,input().split())) for _ in range(n)]
board.append([0]*m)

dx=[1,-1,0,0]
dy=[0,0,1,-1]
check=[]

def bfs(x,y,board):
    queue=deque()
    queue.append((x,y))
    check.append((x,y))
    point=[]
    nQueue=deque()
    distance=0
    while (True):
        while queue:
            x,y=queue.popleft()
            for i in range(4):
                nx,ny=x+dx[i],y+dy[i]
                if (0<=nx<n and 0<=ny<m and (nx,ny) not in check):
                    check.append((nx,ny))
                    if board[nx][ny]==1:
                        point.append((nx,ny))
                    else:
                        nQueue.append((nx,ny))
        distance+=1
        if distance>d:
            return -1,-1
        if point:
            if len(point)>1:
                point=sorted(point,key=lambda x:x[1])
            return point[0][0], point[0][1]
        else:
            queue=nQueue
            nQueue=deque()
            
def remainEnemy(board):
    for i in range(n):
        for j in range(m):
            if board[i][j]==1:
                return False
    return True

    
def killEnemy(board):
    count=0
    while True:
        enemy=set()
        for i in range(m):
            if board[n][i]==2:
                ex,ey=bfs(n,i,board)
                if ex!=-1:
                    enemy.add((ex,ey))
        count+=len(enemy)
        for x,y in enemy:
            board[x][y]=0
            
        for i in range(n-1,-1,-1):
            for j in range(m):
                if board[i][j]==1:
                    board[i][j]==0
                    if i!=n-1:
                        board[i+1][j]=1
        if remainEnemy(board):
            return count
    
result=0
def solution(count, board):
    global result
    if count==3:
        board=list(board)
        result=max(result, killEnemy(board))
        return
    for i in range(m):
        if board[n][i]==0:
            board[n][i]=2
            solution(count+1, board)
            board[n][i]=0
            
solution(0, board)
print(result)
```
