## 3190 뱀

**문제로 이동** : [Baekjoon 뱀](https://www.acmicpc.net/problem/3190)

- 뱀이 사과를 먹으면 길이가 늘어나고, 벽 또는 자기자신과 부딫히면 게임이 끝난다
- 게임은 N\*N 보드 위에서 진행되면 보드의 상하좌우 끝에 벽이 있다
- 게임 시작 시 뱀의 위치는 맨위 맨 좌측에 위치하고, 길이는 1이다
- 뱀은 매 초 이동을 하는데 규칙은 다음과 같다
  1. 먼저 뱀은 몸 길이를 늘려 머리를 다음 칸에 위치시킨다
  2. 이동한 칸에 사과가 있다면 사과가 없어지고 꼬리는 움직이지 않는다
  3. 사과가 없다면 몸길이를 줄여 꼬리가 위치한 칸을 비운다 (몸 길이 변화없다)
- 사과의 위치와 뱀의 이동경로가 주어질 때 게임이 몇 초 안에 끝나는지 계산하라


```python
import sys
from collections import deque

#input=sys.stdin.readline

n=int(input())
board=[[0]*n for _ in range(n)]
k=int(input())
for i in range(k):
    x,y=map(int,input().split())
    board[x-1][y-1]='apple'

moveNum=int(input())
moveQ=deque()
for i in range(moveNum):
    x,c=input().split()
    x=int(x)
    moveQ.append((x,c))
    
dx=[0,1,0,-1]
dy=[1,0,-1,0]
sec=0
direction=0
x,y=0,0
board[x][y]='snake'
snake=[[x,y]]
info=moveQ.popleft()

while (True):
    nx=x+dx[direction]
    ny=y+dy[direction]
    
    if (0<=nx<n and 0<=ny<n and board[nx][ny]!='snake'):
        if (board[nx][ny]==0):
            tempX,tempY=snake.pop(0)
            board[tempX][tempY]=0
        board[nx][ny]='snake'
        snake.append([nx,ny])    
    else:
        sec+=1
        break
    x, y=nx,ny
    sec+=1
    
    if sec in info:
        if info[1]=='D':
            direction=(direction+1)%4
        elif info[1]=='L':
            direction=(direction-1)%4
        if moveQ:
            info=moveQ.popleft()
print(sec)
```

## 14499 주사위 굴리기

**문제로 이동** : [Baekjoon 주사위 굴리기](https://www.acmicpc.net/problem/14499)

- 크기가 N\*M인 지도에서 오른쪽은 동쪽, 위쪽은 북쪽이다
- 지도의 좌표는 (r, c)로 나타내며, r은 북쪽으로 떨어진 개수, c는 서쪽으로 떨어진 칸의 개수이다
- 지도의 각 칸에는 정수가 쓰여져 있고 처음 주사위의 모든면은 0이다
- 주사위를 굴렸을 때 이동한 칸에 쓰여있는 수가 0이면 주사위의 바닥면에 쓰여있는 수가 칸에 복사된다
- 0이 아닌 경우 칸에 쓰여있는 수가 주사위의 바닥면으로 복사되며, 칸은 0이 된다
- 이동할 때마다 주사위의 윗 면에 쓰여있는 수를 출력한다

```python
N, M, x, y, K = map(int, input().split()) 
maps = [list(map(int, input().split())) for _ in range(N)] 
op = list(map(int, input().split())) 
dice = [0]*7

def move(op): 
    if op == 1: #동쪽
        dice[1], dice[3], dice[4], dice[6] = dice[3], dice[6], dice[1], dice[4] 
    elif op == 2: #서쪽
        dice[1], dice[3], dice[4], dice[6] = dice[4], dice[1], dice[6], dice[3] 
    elif op == 3: #북쪽
        dice[1], dice[2], dice[5], dice[6] = dice[2], dice[6], dice[1], dice[5] 
    elif op == 4: #남쪽
        dice[1], dice[2], dice[5], dice[6] = dice[5], dice[1], dice[6], dice[2] 
        
def direction(op): 
    if op == 1: 
        return (0, 1) 
    elif op == 2: 
        return (0, -1) 
    elif op == 3: 
        return (-1, 0) 
    elif op == 4: 
        return (1, 0) 
    
for i in op:  
    moveX, moveY = direction(i) 
    if 0 <= x+moveX < N and 0 <= y+moveY < M: 
        x += moveX 
        y += moveY 
        move(i)
        if maps[x][y] != 0: 
            dice[1] = maps[x][y] 
            maps[x][y] = 0 
        else:
            maps[x][y] = dice[1] 
        print(dice[6])
```

## 14502 연구소

**문제로 이동** : [Baekjoon 연구소](https://www.acmicpc.net/problem/14502)

- 바이러스의 확산을 막기위해 연구소에 벽을 세우려고 한다
- 일부 칸은 바이러스가 존재하며 이 바이러스는 상하좌우로 인접한 빈 칸으로 모두 퍼져나갈 수 있다.
- 새로 세울 수 있는 벽은 3개이며 꼭 3개를 세워야 한다
- 벽을 세운 후 바이러스가 퍼질 수 없는 곳을 안전 영역이라한다
- 연구소의 지도가 주어질 때 얻을 수 있는 안전 영역 크기의 최댓값을 구하라

```python
from itertools import combinations
import sys, copy

#input = sys.stdin.readline
sys.setrecursionlimit(10**7)
n,m = map(int, input().split())
lab = [list(map(int, input().split())) for _ in range(n)]
     

virus, zero=[],[]
for i in range(n):
    for j in range(m):
        if lab[i][j]==2:
            virus.append([i,j])
        elif lab[i][j]==0:
            zero.append([i,j])

dx=[-1,1,0,0]
dy=[0,0,1,-1]
def DFS(x,y):
    wallLab[x][y]=2
    for i in range(4):
        nx = x + dx[i] 
        ny = y + dy[i]
        if (0<=nx<n and 0<=ny<m and wallLab[nx][ny]==0):
            DFS(nx, ny)
            
ableWall=combinations(zero, 3)
safeList=[]
for able in ableWall:
    wallLab=copy.deepcopy(lab)
    wallLab[able[0][0]][able[0][1]]=1
    wallLab[able[1][0]][able[1][1]]=1
    wallLab[able[2][0]][able[2][1]]=1
    
    for v in virus:
        DFS(v[0],v[1])      
    safe=0
    for r in range(n):
        for c in range(m):
            if wallLab[r][c]==0:
                safe+=1
    safeList.append(safe)
    
print(max(safeList))
```

## 10026 적록색약

**문제로 이동** : [Baekjoon 적록색약](https://www.acmicpc.net/problem/10026)

- 크기가 N\*N인 그리드의 각 칸에 RGB 중 하나를 색칠한 그림이 있다
- 같은 색상이 상하좌우로 인접해 있는 경우에 두 글자는 같은 구역에 속한다
- 그림이 입력으로 주어질 때 적록색약인 사람이 봤을 때와 아닌 사람이 봤을 때의 구역 수를 구하는 프로그램을 작성하라

```python
import sys 
from collections import deque 

input=sys.stdin.readline
def BFS(i, j, color, arr): 
    queue = deque() 
    queue.append((i, j))
    arr[i][j] = 0 
    while queue: 
        x, y = queue.popleft() 
        for idx in range(4):
            nx = x + dx[idx] 
            ny = y + dy[idx]
            if (0<=nx<n and 0<=ny<n and arr[nx][ny]==color):
                arr[nx][ny]=0
                queue.append((nx,ny))
                
                
dx = [-1, 1, 0, 0] 
dy = [0, 0, -1, 1] 
n = int(input()) 
board = [list(input()) for _ in range(n)] 

boardBlind = [[0] * n for _ in range(n)]
for i in range(n): 
    for j in range(n): 
        if board[i][j] == 'R' or board[i][j] == 'G': 
            boardBlind[i][j] = 'R' 
        else: 
            boardBlind[i][j] = 'B' 
            
area = 0 
colorBlindArea = 0 
            
for i in range(n): 
    for j in range(n):
        if board[i][j] != 0: 
            BFS(i, j, board[i][j], board) 
            area += 1
            
        if boardBlind[i][j] != 0: 
            BFS(i, j, boardBlind[i][j], boardBlind) 
            colorBlindArea += 1
            
print(area, colorBlindArea)
```
