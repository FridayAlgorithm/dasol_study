## 21608 상어 초등학교

**문제로 이동** : [Baekjoon 상어 초등학교](https://www.acmicpc.net/problem/21608)

- 선생님은 학생의 순서대로 규칙을 이용해 학생의 자리를 정하려고 한다
  * 비어있는 칸 중 좋아하는 학생이 가장 많이 인접한 자리를 정한다
  * 1번 조건을 만족하는 자리가 여러개이면 인접칸이 많은 자리로 정한다
  * 2번 조건을 만족하는 자리가 여러자리면 행의 번호-> 열의 번호가 작은 순으로 정한다
- 학생의 만족도는 인접한 칸에 앉은 좋아하는 학생 수로 구할 수 있다
- 만족도는 좋아하는 학생 수에 따라 0, 1, 10, 100, 1000이다

```python
import sys

# input=sys.stdin.readline

dx=[0,-1,0,1]
dy=[-1,0,1,0]
n=int(input())
classroom=[[0]*n for _ in range(n)]
likeStudent={}
result=0

for _ in range(n*n):
    student=list(map(int,input().split()))
    likeStudent[student[0]]=student[1:]
    sNum=student[0]
    maxX,maxY=0,0
    maxLike, maxEmpty=-1,-1
    for i in range(n):
        for j in range(n):
            likeCnt, emptyCnt=0,0
            if classroom[i][j]==0:
                likeCnt, emptyCnt=0,0
                for k in range(4):
                    nx=i+dx[k]
                    ny=j+dy[k]
                    if (0<=nx<n and 0<=ny<n):
                        if classroom[nx][ny] in likeStudent[sNum]:
                            likeCnt+=1
                        if classroom[nx][ny]==0:
                            emptyCnt+=1
                if maxLike<likeCnt or (maxLike==likeCnt and maxEmpty<emptyCnt):
                    maxX, maxY=i,j
                    maxLike, maxEmpty=likeCnt, emptyCnt
    classroom[maxX][maxY]=sNum
    
for i in range(n):
    for j in range(n):
        count=0
        for k in range(4):
            nx=i+dx[k]
            ny=j+dy[k]
            if (0<=nx<n and 0<=ny<n):
                if classroom[nx][ny] in likeStudent[classroom[i][j]]:
                    count+=1
        if count!=0:
            result+=10**(count-1)
            
print(result)
```

## 14888 연산자 끼워넣기

**문제로 이동** : [Baekjoon 연산자 끼워넣기](https://www.acmicpc.net/problem/14888)

- 수와 수 사이에 연산자를 넣어서 수식을 만들 수 있다
- 식의 계산은 연산자 우선순위를 무시하고 앞에서부터 진행된다
- N개의 수와 N-1개의 연산자가 주어졌을 때 결과의 최댓값과 최솟값을 구하라

```python
import sys

# input = sys.stdin.readline
n=int(input())
number=list(map(int,input().split()))
add, sub, mul, div=map(int,input().split())
maxNum, minNum=-1000000000,1000000000

def dfs(i, res, add, sub, mul, div):
    global maxNum, minNum
    if i==n:
        if res>maxNum:
            maxNum=res
        if res<minNum:
            minNum=res
        return
    else:
        if add:
            dfs(i+1, res+number[i], add-1, sub, mul, div)
        if sub:
            dfs(i+1, res-number[i], add, sub-1, mul, div)
        if mul:
            dfs(i+1, res*number[i], add, sub, mul-1, div)
        if div:
            dfs(i+1, int(res/number[i]), add, sub, mul, div-1)
            
dfs(1,number[0],add,sub,mul,div)
print(maxNum)
print(minNum)
```

## 17070 파이프 옮기기 1

**문제로 이동** : [Baekjoon 파이프 옮기기1](https://www.acmicpc.net/problem/17070)

- 유현이는 집 수리를 위해 파이프를 옮기고자 한다
- 파이프는 회전이 가능하며 3가지 방향이 가능하다
- 회전은 45도만 회전할 수 있으며 미는 방향은 오른쪽, 아래, 대각선이다
- 처음 파이프는 (1,1)와 (1,2)를 차지하고 있고 방향은 가로이다
- 파이프의 한쪽 끝을 (N,N)으로 이동시키는 방법의 수를 구하라

```python
import sys 

# input=sys.stdin.readline
directions = {0: [0, 2], 1: [1, 2], 2: [0, 1, 2]} # 가로(0) / 세로(1) / 대각선(2) 
cos = {0: [0, 1], 1: [1, 0], 2: [1, 1]} # (y, x) 
n = int(input()) 
house = [] 
for _ in range(n): 
    house.append([int(x) for x in input().split()]) 

def check(y, x, d): 
    for direction in directions[d]: 
        dy, dx = cos[direction] 
        ny = y + dy 
        nx = x + dx 
        if 0 <= ny < n and 0 <= nx < n and not house[ny][nx]: 
            if direction != 2:
                dp[ny][nx][direction] += dp[y][x][d] 
            else:
                if not house[ny-1][nx] and not house[ny][nx-1]: 
                    dp[ny][nx][direction] += dp[y][x][d] 
                    
dp = [[[0 for _ in range(3)] for _ in range(n)] for _ in range(n)] 
dp[0][1][0] = 1 
for y in range(n): 
    for x in range(n): 
        for d in range(3):
            if dp[y][x][d] and not house[y][x]: 
                check(y, x, d) 
                
print(sum(dp[n-1][n-1]))
```

## 14501 퇴사

**문제로 이동** : [Baekjoon 퇴사](https://www.acmicpc.net/problem/14501)

- 오늘부터 N+1일째 되는 날 퇴사를 하기 위해 N일동안 최대한 많은 상담을 하려고 한다
- 각 상담은 걸리는 시간과 받을 수 있는 금액으로 이루어져있다
- 상담을 적절히 했을 때 얻을 수 있는 최대 수익을 구하라

```python
n=int(input())
consulting=[list(map(int,input().split())) for _ in range(n)]
dp=[0 for _ in range(n+1)]

for i in range(n-1,-1,-1):
    if i+consulting[i][0]>n:
        dp[i]=dp[i+1]
    else:
        dp[i]=max(consulting[i][1]+dp[i+consulting[i][0]], dp[i+1])
        
print(dp[0])
```

## 17136 색종이 붙이기

**문제로 이동** : [Baekjoon 색종이 붙이기](https://www.acmicpc.net/problem/17136)

- 1~5크기의 정사각형 모양 색종이를 5개씩 가지고 있다
- 색종이를 크기가 10\*10인 종이위에 붙이고자 한다
- 1이 적힌 칸은 모두 색종이로 덮여져야하며 0이 적힌 칸은 붙여서는 안된다
- 색종이를 붙일 때 경계밖으로 나가거나 겹쳐져서도 안된다
- 종이가 주어졌을 때 1이 적힌 모든 칸을 붙이는데 필요한 색종이의 최소 수를 구하여라

```python
#시간초과 -> pypy 제출
import sys

#input=sys.stdin.readline
            
def check(i,j,size,board):
    target = []
    for y in range(size) :
        for x in range(size) :
            ny, nx = i+y, j+x
            if ny < 0 or ny >= 10 or nx < 0 or nx >= 10 :
                return False, []
            if board[ny][nx] == 0 :
                return False, []
            else :
                target.append((ny, nx))

    temp = [[board[y][x] for x in range(10)] for y in range(10)]
    for y,x in target:
        temp[y][x] = 0
    return True, temp


def DFS(board, remain, papers):
    global result
    if remain == 0:
        result = min(result,25 - sum(papers))
        return
    for i in range(10):
        for j in range(10):
            if board[i][j] == 1:
                flag = False
                for size in range(1,6):
                    if papers[size] > 0:
                        checkRes, temp = check(i,j,size,board)
                        if checkRes:
                            papers[size] -= 1
                            DFS(temp,remain-(size**2),papers)
                            papers[size] += 1
                        flag = True
                if flag:
                    return

                
board = [list(map(int,input().split())) for _ in range(10)]
remain = 0
for i in range(10):
    for j in range(10):
        if board[i][j] == 1:
            remain += 1

papers = [0,5,5,5,5,5]
result = 26

DFS(board, remain, papers)
if result == 26:
    print(-1)
else:
    print(result)
```
