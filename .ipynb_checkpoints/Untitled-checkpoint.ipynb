## 전산물리 과제

- x,y의 범위 (9)와 문제 번호(9999)를 사용자로부터 입력받는다
   * 범위 안맞는 수가 들어오면 오류메시지와 함께 다시 입력
      - 범위를 2부터 9로 제한해 주세요.
   * 문제 번호 형식과 맞지 않을 시 오류메시지와 함께 다시 출력
      - 문제 번호 형식이 잘못 되어 있습니다
- 문제 번호에는 좌표가 저장 (random으로 seed값 설정)
   * 0 들어오면 랜덤으로 문제번호 지정
- 알파벳 들어오면 오류메시지 생성하고 다시 입력받도록
- 변환할 x,y좌표를 띄어쓰기 없이 입력받는다
- 입력받은 좌표의 상하좌우를 1->0 0->1 변환 후 출력
- 0을 입력받을 때까지 계속 좌표변환 진행 (0 입력시 종료)
- 모든 board의 값이 0 or 1이 되면 게임 종료


```python
import random

def makeRandomBoard(m, n):
    board=[]
    for i in range(n):
        temp=[]
        for j in range(m):
            temp.append(random.randint(0,1))
        board.append(temp)
    return board
    

problemBoard=[0]*9999
x=-1; y=-1; problemNum=-1
while (x<2 or x>9):
    x=input("x축의 크기 (2-9)? ")
    if (x.isalpha()):
        #숫자가 아닌 것 입력받을 시
        print("숫자를 입력해주세요"); x=0
    else:
        x=int(x)
        if (x<2 or x>9):
            print("범위를 2부터 9로 제한해 주세요")
            
while (y<2 or y>9):
    y=input("y축의 크기 (2-9)? ")
    if (not y.isdigit()):
        #숫자가 아닌 것 입력받을 시
        print("숫자를 입력해주세요"); y=0
    else:
        y=int(y)
        if (y<2 or y>9):
            print("범위를 2부터 9로 제한해 주세요")


while (problemNum<1 or problemNum>9999):
    problemNum=input("문제 번호는 (1-9999)? ")
    if (not problemNum.isdigit()):
        print("문제 번호 형식이 잘못되어 있습니다"); problemNum=-1
    else:
        problemNum=int(problemNum)
        if (problemNum<0 or problemNum>9999):
            print("문제 번호 형식이 잘못되어 있습니다"); problemNum=-1
        elif (problemNum == 0):
            problemNum=random.randint(0, 9999)
        
if (problemBoard[problemNum]==0):
    board=makeRandomBoard(x,y)
    problemBoard[problemNum]=board

print(problemNum,"번 문제입니다.")
board=problemBoard[problemNum]
for i in board:
    for j in i:
        print(j, end=' ')
    print()

while (True):
    print("xy좌표(두자리 수)는? (종료를 원하면 0을 넣으세요)")
    change=list(input()) #[x, y]
    if (int(change[0])==0):
        print("종료합니다.")
        break
    a=int(change[0]); b=int(change[2])
    changeX,changeY=x-int(b)-1,int(a)-1 #좌표 인덱스 계산
    #x,y값이 다를 때 인덱스 계산
    
    #board변환
    board[changeX][changeY]=int(not board[changeX][changeY])
    if (changeX+1<x): #아래
        board[changeX+1][changeY]=int(not board[changeX+1][changeY])
    if (changeY+1<y): #오른쪽
        board[changeX][changeY+1]=int(not board[changeX][changeY+1])
    if (changeX-1>=0): #위
        board[changeX-1][changeY]=int(not board[changeX-1][changeY])
    if (changeY-1>=0): #왼쪽
        board[changeX][changeY-1]=int(not board[changeX][changeY-1])
        
    if (sum(board)==0):
        print("승리했습니다!")

    for i in board:
        for j in i:
            print(j, end=' ')
        print()

```

# 시그마

- bisection method (이분법)
  * 구간 [a, b]에서 연속하는 함수에 대해 f(a) * f(b)<0이면 f(x)의 해는 구간 사이에 존재한다
  

```python
import math
def g(k,x,i):
    global a
    a=1
    for k in range(1,i+1):
        a*=(k**0.3-x)/(k**0.3+x)
    return a

def f(x) :
    b=0
    for i in range(1,1001) :
        
        b+=g(k,x,i)
    return b

err=1e-9
a1=0
b1=2
error=b1-a1
iteration=0

while error > err:
    m = (a1+b1)/2
    if f(m)>1:
        a1=m
    else:
        b1=m
    error=abs(b1-a1)/m
    iteration+=1
print(m,iteration)

print(f(1))
print(f(0))
```

# 야구게임

- 1~9까지의 숫자를 난수로 겹치지 않게 입력받는다
- 백의자리, 십의자리, 일의자리 숫자 3개의 숫자를 입력받는다
- 숫자와 자릿수가 모두 같은 때는 STRIKE, 숫자만 같고 자리수가 다를 경우 BALL이다
- 10번의 기회 안에 숫자를 맞출 경우 승리, 그렇지 못할 경우 패배를 출력한다
- 승패와 상관없이 게임이 끝나면 정답을 출력한다

```python
import random

def MakeRandomNum():
    ranList=[1,2,3,4,5,6,7,8,9]
    randomNum=random.sample(ranList, 3)
    return randomNum

if __name__=="__main__":
    answer=MakeRandomNum()
    flag=False
    for _ in range(10):
        inputNum=list(map(int,input().strip()))
        if (answer==inputNum):
            flag=True
            break
        strike, ball=0,0
        for i in range(3):
            if (inputNum[i]==answer[i]):
                strike+=1
            elif (inputNum[i] in answer):
                ball+=1
        if (strike==3):
            flag=True
            break
        else:
            print(strike,"스트라이크 ",ball,"볼")
    if flag:
        print("승리했습니다!")
    else:
        print("패배했습니다..")
    print("정답은 ", *answer, sep='')
```
